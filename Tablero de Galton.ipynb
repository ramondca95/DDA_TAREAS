```python

from random import randrange
import random
import numpy as np
from matplotlib import pyplot as plt
import pandas
from collections import Counter

def tablero_galton (numBolitas, niveles,inciso):
  
  x = 0
  lista2 = []
  list_hist = []
  if inciso == 1:
    sesgoIzquierda = 50
  if inciso == 2: 
    sesgoIzquierda = 60
  
  for j in range(1,numBolitas+1):
    lista = []  
    for i in range(1,niveles+1):
      if inciso == 3:
        sesgoIzquierda = 50 + (50/niveles)*i
        if sesgoIzquierda > 100:
          sesgoIzquierda = 100
        
      x = random.randrange(1,101,step = 1)
      if x < sesgoIzquierda:
        x = 0
      else:
        x = 1
      lista.append(x)
      
    lista2.append(sum(lista))
    lista2.sort()
  listFrecuencia = Counter(lista2)
  df = pandas.DataFrame.from_dict(listFrecuencia, orient = 'index')
  df.plot(kind='bar', figsize=(15,10),title='INCISO ' + str(inciso) + ': ' + str(numBolitas) + ' bolitas y '+ str(niveles)+ ' niveles')
```

```python
tablero_galton(100,20,1)
tablero_galton(1000,30,1)
tablero_galton(10000,50,1)
```

```python
tablero_galton(100,20,2)
tablero_galton(1000,30,2)
tablero_galton(10000,50,2)
```
