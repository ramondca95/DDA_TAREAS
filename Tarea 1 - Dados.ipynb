## TAREA: OBTENER LA ECUACIÓN QUE EXPRESE LA PROBABILIDAD DE OBTENER EL MACROESTADO **E** A PARTIR DE **N** DADOS.

Para comprender la tarea es importante definir qué es un macroestado y un microestado para este problema en específico.

El microestado(MiE) sería la cara de los dados que, cuando son dos o más dados el conjunto de microestados conforman un macroestado(MaE).

Ej. Microestado: la cara con el número 3.

Ej. Macroestado de dos dados: la suma de la cara 3 y la cara 4 para formar el MaE 7. Mismo MaE para la suma de la cara 2 y 5 o 6 y 1.



Para esta tarea, es necesario encontrar el número de combinaciones con repetición que generen el macroestado **E** y dividirla entre el total de combinaciones con repetición que se puedan generar. 



```python
import itertools
from itertools import groupby
import pandas
from collections import Counter
from fractions import Fraction

dado = [1,2,3,4,5,6]

def prob_macroestado(cantidadDados,macroestado,printCombinaciones = False):
  """ Función para encontar la probabilidad de que ocurra un macroevento cuando se lanzan n dados"""
  sumaCombinaciones = []
  listProbabilidades = []
  listFrecuencia = []
  listValores = []

  """ En una lista se almacenan todas las posibles combinaciones de microestados posibles al arrojar n cantidad de dados, 
  la longitud de esta lista responde a la función Número de Caras ^ Cantidad de dados."""

  list1 = [p for p in itertools.product(dado, repeat=cantidadDados)]

  """ Se recorre la lista para almacenar el resultado de sumar las combinaciones de los microestados, es decir que 
  almacenará los macroestados. Con la condicional if se almacenan en otra lista las combinaciones que al sumarse 
  el resultado sea el valor buscado valorDados."""

  for i in range(0,len(list1)):
    sumaCombinaciones.append(sum(list1[i]))
    if sum(list1[i]) == macroestado:
      listValores.append(list1[i])
  contador = len(listValores)
   
  """ Con la función COUNTER se cuentan las veces que aparece un valor en una en la lista """
  listFrecuencia = Counter(sumaCombinaciones)
  
  
  
  if printCombinaciones == True:
    print('\nCombinaciones: ',listValores, '\n')
    df = pandas.DataFrame.from_dict(listFrecuencia, orient = 'index')
    df.plot(kind='bar', figsize=(15,10))
  if macroestado == None:
    df = pandas.DataFrame.from_dict(listFrecuencia, orient = 'index')
    df.plot(kind='bar', figsize=(15,10))
    print("El macroestado más común: ",listFrecuencia.most_common(1)[0][0])
  else:
    """ Se muestra el resultado, el histograma y las combinaciones solo si se declara True"""
    print("\nLa probabilidad de que ", macroestado, " salga cuando se tiran ", cantidadDados, "dados es: ", Fraction(contador,len(sumaCombinaciones))," = ", contador/len(sumaCombinaciones))
  
```

```python
prob_macroestado(2,7, True)
```

```python
cantidadDados = 2
for k in range(cantidadDados,6*cantidadDados + 1):
  prob_macroestado(cantidadDados, k)
```
